```python
from google.colab import userdata
openai_api_key=userdata.get('OPENAI_API_KEY')
from openai import OpenAI
client=OpenAI(api_key=openai_api_key)
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large", openai_api_key=openai_api_key)

texts = [
    "Car is a vehicle",
    "Automobile is a vehicle",
    "Apple is a fruit"
]

vectors = embeddings.embed_documents(texts)

print("\nEmbeddings Generated\n")

for i, vector in enumerate(vectors):
    print(f"Text: {texts[i]}")
    print(f"Vector (first 5 values): {vector[:5]}")
    print("-" * 50)
```

```python
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
vectors = embeddings.embed_documents(texts)

print("\nSimilarity:\n")

print("Car vs Automobile:",
      cosine_similarity(vectors[0], vectors[1]))

print("Car vs Apple:",
      cosine_similarity(vectors[0], vectors[2]))
```

```python
from pypdf import PdfReader
from langchain_openai import OpenAIEmbeddings

# Initialize embedding model
embeddings = OpenAIEmbeddings(openai_api_key=openai_api_key)

reader = PdfReader('/content/sample_data/Github.pdf')
page = reader.pages[0]

text = page.extract_text()

# IMPORTANT: pass as list
vectors = embeddings.embed_documents([text])

print("\nEmbeddings Generated for PDF\n")

print("Text (first 200 chars):")
print(text[:200])

print("\nVector (first 5 values):")
print(vectors[0][:5])
```
