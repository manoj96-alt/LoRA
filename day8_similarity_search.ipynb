```python
from pypdf import PdfReader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
import numpy as np

# Step 1: Load document
reader = PdfReader("/content/sample_data/Github.pdf")

text = ""
for page in reader.pages:
    text += page.extract_text()

print("\nDocument Loaded\n")

# Step 2: Chunk the document
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_text(text)

print(f"Total chunks: {len(chunks)}\n")

# Step 3: Create embeddings
embeddings = OpenAIEmbeddings(openai_api_key=openai_api_key)

chunk_vectors = embeddings.embed_documents(chunks)

# Step 4: Define similarity function
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# Step 5: Ask user query
query = input("Ask a question: ")

query_vector = embeddings.embed_query(query)

# Step 6: Compare with all chunks
similarities = []

for i, vector in enumerate(chunk_vectors):
    score = cosine_similarity(query_vector, vector)
    similarities.append((i, score))

# Step 7: Sort by highest similarity
similarities = sorted(similarities, key=lambda x: x[1], reverse=True)

# Step 8: Get best match
best_match_index = similarities[0][0]

print("\nMost Relevant Chunk:\n")
print(chunks[best_match_index][:500])
```
