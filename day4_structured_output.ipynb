```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
from openai import OpenAI
client=OpenAI(api_key=OPENAI_API_KEY)
```

```python

import json



print("Structured AI Output Example\n")

question = input("Enter a concept to explain: ")

prompt = f"""
You are an AI tutor.

Return the answer ONLY in JSON format.

Example format:

{{
 "concept": "",
 "definition": "",
 "example": ""
}}

Concept: {question}
"""

response = client.responses.create(
    model="gpt-4.1-mini",
    input=prompt
)

output = response.output_text

print("\nRaw AI Output:\n")
print(output)

# Convert JSON string to Python dictionary
try:
    data = json.loads(output)

    print("\nParsed JSON:\n")
    print("Concept:", data["concept"])
    print("Definition:", data["definition"])
    print("Example:", data["example"])

except:
    print("\nCould not parse JSON. Try again.")
```

```python

import json



print("Structured AI Output Example\n")

question = input("Enter a concept to explain: ")

prompt = f"""
You are an AI tutor.

Return the answer ONLY in JSON format.

Example format:

{{
 "concept": "",
 "definition": "",
 "example": ""
 "difficulty_level":""
}}

Concept: {question}
"""

response = client.responses.create(
    model="gpt-4.1-mini",
    input=prompt
)

output = response.output_text

print("\nRaw AI Output:\n")
print(output)

# Convert JSON string to Python dictionary
try:
    data = json.loads(output)

    print("\nParsed JSON:\n")
    print("Concept:", data["concept"])
    print("Definition:", data["definition"])
    print("Example:", data["example"])
    print("Difficulty Level:", data["difficulty_level"])

except:
    print("\nCould not parse JSON. Try again.")
```
