```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
response=client.responses.create(model="gpt-3.5-turbo",
                                 input="explain the concept of a large language model in simple terms"
)
print(response.output_text)
```

```python
response=client.responses.create(model="gpt-3.5-turbo",input="Explain what embeddings are in simple terms")
print(response.output_text)
```

```python
response=client.responses.create(model="gpt-3.5-turbo",input="Explain RAG in 3 bullet points.")
print(response.output_text)
```

```python
question=input("Ask a question:")
response=client.responses.create(model="gpt-4.1-mini",input=question)
print(response.output_text)
```
