```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python

print("AI Chatbot (type 'exit' to stop)\n")

while True:
  user_input = input("You: ")

  if user_input.lower() == "exit":
    print("Goodbye!")
    break
  response = client.responses.create(
        model="gpt-4.1-mini",
        input=user_input
  )

  print("\nAI:", response.output_text)
  print()
```

```python

print("AI Chatbot (type 'exit' to stop)\n")

while True:
  user_input = input("You: ")
  prompt=f"You are a helpful AI tutor. Answer clearly.\n\nQuestion: {user_input}"

  if user_input.lower() == "exit":
    print("Goodbye!")
    break
  response = client.responses.create(
        model="gpt-4.1-mini",
        input=prompt
  )

  print("\nAI:", response.output_text)
  print()
```

```python

print("AI Chatbot (type 'exit' to stop)\n")

conversation = ""

while True:

    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Goodbye!")
        break

    # Add user message to conversation history
    conversation += f"\nUser: {user_input}"

    prompt = f"""
You are a helpful AI assistant.

Here is the conversation so far:

{conversation}

AI:
"""

    response = client.responses.create(
        model="gpt-4.1-mini",
        input=prompt
    )

    ai_reply = response.output_text.strip()

    print("\nAI:", ai_reply)
    print()

    # Add AI response to conversation history
    conversation += f"\nAI: {ai_reply}"
```
